```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
```

```python
sales_per_day = pd.read_csv(
    'C:\\netology\\net2020\\matplot\\3.csv',
    sep=';'
    ,decimal=','
    ,names = ['Date','Rev','Rev_PY','Продажи','Продажи прошлый год']
    )
first_date = sales_per_day['Date'].min()[:10]
last_date = sales_per_day['Date'].max()[:10]    

plot = sales_per_day.set_index(['Date'])[['Продажи','Продажи прошлый год']].plot(rot=45)

plot.set_title(f'Продажи накопительно c {first_date} по {last_date}')
```

```python
sales_per_tp = pd.read_csv(
    'C:\\netology\\net2020\\matplot\\2.csv',
    sep=';'
    ,decimal=','
    ,names = ['Date','TradePoint','Revenue']
    
    )
sales_per_tp

ax = sns.boxplot(x="TradePoint", y="Revenue", orient = "v", 
            data=sales_per_tp)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45,horizontalalignment='right',fontweight='light',fontsize='large')
ax.set_title('Средние продажи в день по ТОП торговым точкам')
```
